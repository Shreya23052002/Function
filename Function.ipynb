Question no. 1 What is the difference between a function and a method in Python?

Answer :
 ### Function:
A function is a block of code that can be defined independently, outside of any class. It can take inputs (parameters), perform a task, and optionally return an output.
You define a function using the def keyword.

```python
# Example
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

### Method:
A method is essentially a function that is associated with an object (i.e., a class instance). Methods are defined inside classes and are used to operate on data that belongs to the class or object.

There are two main types of methods: instance methods and class methods.

Instance methods take the instance (self) as their first parameter, allowing them to access and modify the attributes of that particular object.

```python
# Example:
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

person = Person("Alice")
print(person.greet())
```

An iterable is any Python object that can return an iterator. It is essentially a collection of elements that can be iterated over, such as:

An iterable must implement the __iter__() method, which returns an iterator.

# What is an Iterator?
An iterator is an object that remembers its state during iteration. It must implement two methods:

- __iter__() â†’ Returns the iterator object itself.

- __next__() â†’ Returns the next item in the sequence and raises StopIteration when no items are left.

```python
# Example:
my_iterator = iter(my_list)  # Converting a list into an iterator

print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
print(next(my_iterator))  # Raises StopIteration
```

Question no. 6 Explain the concept of generators in Python and how they are defined.

Answer:
### Generators in Python
Generators are a special type of iterator in Python that allow lazy evaluation, meaning they generate values on the fly rather than storing them in memory. They are useful for working with large datasets and infinite sequences.
### How Generators Work
Generators are created using functions with the yield keyword instead of return. Unlike regular functions, generators pause execution when they yield a value and resume from where they left off when called again.

- Key Characteristics of Generators
- Use yield instead of return

- Automatically implement __iter__() and __next__()

- Generate values lazily (one at a time, as needed)

- More memory-efficient than lists (do not store all elements at once)

Question no 9 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Answer: These three functionsâ€”map(), filter(), and reduce()â€”are part of functional programming in Python. They allow operations on iterables in a concise and efficient way.
### map() â€“ Apply a Function to Each Element:
Applies a given function to each element of an iterable and returns a new iterable with the modified value

```python
# Example:
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16]
```

### filter() â€“ Keep Only Matching Elements:
Filters elements based on a condition (function returning True or False).

```python
# Example:
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]
```

### reduce() â€“ Apply Function Cumulatively:
Reduces an iterable to a single value by applying a function cumulatively.

```python
# Example:
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24 (1 * 2 * 3 * 4)
```

Question no. 1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    # Use filter() to select only even numbers and then sum them up
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    return sum(even_numbers)

# Example usage:
numbers = [47, 11, 42, 13, 8, 56]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)  # Output: 106 (42 + 8 + 56)
```

Question no. 4 Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    # Handle edge cases for numbers less than 2
    if num < 2:
        return False
    # Check divisibility from 2 to the square root of num
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Example usage:
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is prime.")
```

Question no. 5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n  # Number of terms
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # To keep track of the number of terms generated

    def __iter__(self):
        return self  # The class itself is the iterator

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # Stop iteration when the specified number of terms is reached

        # Return the next Fibonacci number
        fib_number = self.a
        self.a, self.b = self.b, self.a + self.b  # Update the values of a and b to the next Fibonacci numbers
        self.count += 1
        return fib_number

# Example usage:
n_terms = 10
fib_iterator = FibonacciIterator(n_terms)

for num in fib_iterator:
    print(num)
```

Question no. 6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
exponent = 5
for power in powers_of_two(exponent):
    print(power)
```

Question no. 9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40, 50]

# Use map() to apply the conversion function to each temperature in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the result
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

Question no. 10  Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Function to filter out vowels from a string
def remove_vowels(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Input string
input_string = "Hello, World!"

# Use filter() to remove vowels by applying the remove_vowels function
filtered_string = ''.join(filter(remove_vowels, input_string))

# Print the result
print("String without vowels:", filtered_string)
```
